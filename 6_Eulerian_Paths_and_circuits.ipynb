# Eulerian Paths and Eulerian Circuits

* Eulerian paths is the **path of edges** that **visits all the edges** in a graph exactly once, it depends of wich is the node of start
* Eulerian circuit is an eulerian path **wich starts and end  on the same vertex**

### Important requirement for eulerian path/circuits:

* ***All the vertices with non-zero degree need to belong to a single connected component***

## Properties to the existence of eulerian paths or circuits



## Example



```python
class Node():
    def __init__(self, id_num):
        self.num = id_num
        self.frm_nodes = []
        self.to_nodes = []
    
    def add_to(self, to_input):
        self.to_nodes.append(to_input)
    
    def add_frm(self, frm_input):
        self.frm_nodes.append(frm_input)
    
class graph():

    def __init__(self):
        self.total_nodes = 0
        self.nodes = {}

    def add_node(self, node):
        self.nodes[node.num] = node
        self.total_nodes += 1

def Eulerian_path (graph, number_edges):
    in_array = [0] * graph.total_nodes
    out_array = [0] * graph.total_nodes
    path = []

    def count_in_out_degrees():

        for node in graph.nodes.keys():
            out_array[graph.nodes[node].num] = len(graph.nodes[node].frm_nodes)
            in_array[graph.nodes[node].num] = len(graph.nodes[node].to_nodes)

    def graph_has_eulerian_path():
        start_nodes = 0
        end_nodes = 0

        for i in range(graph.total_nodes):
            if (out_array[i] - in_array[i]) > 1 or (in_array[i] - out_array[i]) > 1:
                return False
            elif (out_array[i] - in_array[i]) == 1:
                start_nodes += 1
            elif (in_array[i] - out_array[i]) == 1:
                start_nodes += 1

        if (start_nodes == 0 and end_nodes == 0) or (start_nodes == 1 and end_nodes == 1):
            return True

    def find_start_node():
        start = 0

        for i in range(graph.total_nodes):

            if (out_array[i] - in_array[i]) == 1:
                return i

            if out_array[i] > 0:
                start = i

        return start

    def DFS(at):

        while (out_array[at] != 0):

            next_node = graph.nodes[at].frm_nodes.pop()
            out_array[at] -= 1
            DFS(next_node)
        #Error append\n",
        path.append(at)

    count_in_out_degrees()
    if not graph_has_eulerian_path:
        return None
    DFS(find_start_node())

    if len(path) == number_edges +1:
        return path
    else:
        return None
                    

Graph = graph()

node_0 = Node(0)
node_1 = Node(1)
node_2 = Node(2)
node_3 = Node(3)
node_4 = Node(4)
node_5 = Node(5)
node_6 = Node(6)
    
node_1.add_to(1)
node_1.add_to(3)
node_1.add_frm(3)
    
node_2.add_to(4)
node_2.add_to(4)
node_2.add_to(2)
node_2.add_frm(1)
node_2.add_frm(2)
node_2.add_frm(3)
    
node_3.add_to(1)
node_3.add_to(2)
node_3.add_to(5)
node_3.add_frm(6)
node_3.add_frm(4)
node_3.add_frm(1)
    
node_4.add_to(6)
node_4.add_to(3)
node_4.add_frm(2)
node_4.add_frm(2)
    
node_5.add_to(6)
node_5.add_frm(3)
    
node_6.add_to(3)
node_6.add_frm(4)
node_6.add_frm(5)
    
Graph.add_node(node_0)
Graph.add_node(node_1)
Graph.add_node(node_2)
Graph.add_node(node_3)
Graph.add_node(node_4)
Graph.add_node(node_5)
Graph.add_node(node_6)
    
Eulerian_path(Graph, 12)
   
```
