# Bridges in the graph

For find the bridge nodes in the graph **id(e.from) < low (e.to)**

```python
#time solucionado
#corssedge por solucionar
#longitud de lista por solucionar



from collections import defaultdict
import math

class Graph:
    
    def __init__(self, nodes):
        self.neighbors = defaultdict(list)
        self.num_of_nodes = nodes
        self.time = 0
        
    def add_edge(self, frm_node, to_node):
        self.neighbors[frm_node].append(to_node)

    
    def Tarjan_algorithm(self, node_start, discovery, lowest, stack, st):
        
        discovery[node_start] = self.time
        lowest[node_start] = self.time
        stack[node_start] = True
        self.time +=1




            
        for neighbor in self.neighbors[node_start]:

            if discovery[neighbor] == math.inf:

                self.Tarjan_algorithm( neighbor, discovery, lowest,stack,st)
                lowest[node_start] = min(lowest[node_start], lowest[neighbor])

            if stack[neighbor] == True:
                
                lowest[node_start] = min(lowest[node_start], discovery[neighbor])

        # With this if not, i ensure that if in the search the node
        # Doesnt have more neighbors, it means that the DFS return to the parent node
        # and i eliminate this node from the stack, in this ways solved the coss edge problems
        stack[node_start] = False
        
        return lowest
                
    def Call_Tarjan_algorithm (self, node_start_call):
    
        disc = [math.inf] * self.num_of_nodes 
        print(disc)
        low = [math.inf] * self.num_of_nodes
        stack = [False] * self.num_of_nodes
        list_scc = []
        
        print(self.Tarjan_algorithm(node_start_call, disc, low, stack, list_scc))
        


        


    
```
