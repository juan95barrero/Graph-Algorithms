**This is one of the most important algorithm for finding the shortest path on a graph**

* This is a sssp (Single Source Shortest Path) for graph with **Non-negative edge weightes**

**First we define de class node, this class have the following characteristics:**
* parameters : id (integer), adjacent (dictionary), distance (integer), visited(boolean), previous(id of parent)
* add_neighbor : create a pair key, value in the dictionary adjacent, this have two parameters (id of neighbor, weight to that neighbor
* get_conections : call the keys (all the neighbors) from de dictionary adjacent
* get_id : call the integer of it id
* get_weight : it needs the neighbour id to find the weight for that path
* set_distance : set the distance from start node (s) to this specific node
* get_distance : call the distance previously set
* set_previous : this function set the parent node


```python
import math
from queue import PriorityQueue

class Node:
    def __init__(self, node):
        self.id = node
        self.adjacent = {}
        # Set distance to infinity for all nodes
        self.distance = math.inf
        # Mark all nodes unvisited
        self.visited = False  
        # Predecessor
        self.previous = None
    
    def add_neighbor(self, neighbor, weight=0):
        self.adjacent[neighbor] = weight
```

**Second we define de class graph, this class have the following characteristics:**
* parameters : nodes_dict (Dictionary), num_vertices (integer)
* add_node : sum 1 to num_vertices, if it exist, dont do anything, if not it creates and add the object node to the dictionary
* get_node : take the id, and call the object node from the dictionary and return it
* add_edge : take the 'From node', the 'to node' and verify if this exist in the dictionary, if not, it create the nodes and create the 'edge or bridge' between them and set the weight of that path between them, if i want to create a DAG (direct acyclic graph) i need to delete the return path, the weight is only in one way.

**The function shortest return the id's  shortest path from node v to the start**

```python
def shortest(v, path):
    '''make shortest path from v.previous'''
    if v.previous:
        path.append(v.previous.id)
        shortest(v.previous, path)
    return path
    
```

**The dijkstra algorithm 1**
* this algorithm return a list of minimum cumulative path for each node, the shortest path to each node represented in the sum of weights

```python
#Runs Dijkstra Algorithm and returns an array that contains
#The shortest distance to every node from the start node s

# g = graph object
# n = the number of nodes in the graph 
# s = the index of the starting node (0 <= s < n)

def Dijkstra_function (graph, s):

    distance=[math.inf] * graph.num_vertices
    distance[s] = 0
    pq = PriorityQueue()
    pq.put((s , 0))
    
    while pq.empty() == False:
        
        node_id, minValue = pq.get()
        graph.nodes_dict[node_id].visited = True

        for node in graph.nodes_dict[node_id].adjacent.keys():
            
            #if graph.nodes_dict[node].visited == True:
            #    continue
            #else:
            
            new_distance = distance[node_id] + graph.nodes_dict[node_id].adjacent[node]
            if new_distance < distance[node]:
                distance[node] = new_distance
                pq.put((node, new_distance))
                graph.nodes_dict[node].previous = graph.nodes_dict[node_id]

    return distance
    
    
```

```python
class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = defaultdict(list)
        self.distances = {}

    def add_node(self, value):
        self.nodes.add(value)

    def add_edge(self, from_node, to_node, distance):
        self.edges[from_node].append(to_node)
        self.edges[to_node].append(from_node)
        self.distances[(from_node, to_node)] = distance


def dijsktra(graph, initial):
    visited = {initial: 0}
    path = {}
    nodes = set(graph.nodes)
    
    while nodes: 
        min_node = None
        for node in nodes:
            
            if node in visited:
                continue

            if min_node is None:
                min_node = node
            elif visited[node] < visited[min_node]:
                min_node = node

        if min_node is None:
            break

    nodes.remove(min_node)
    current_weight = visited[min_node]

    for edge in graph.edges[min_node]:
        weight = current_weight + graph.distances[(min_node, edge)]
        if edge not in visited or weight < visited[edge]:
            visited[edge] = weight
            path[edge] = min_node

    return visited, path
```
